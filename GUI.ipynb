```python
#Load model
mlp = load_model("bestmodel")
sc = joblib.load("std_scaler")
```

```python
global df

class GUI(tk.Tk):
    def __init__(self):
        tk.Tk.__init__(self)
        self.title('Smart Home Energy Consumption Prediction')
        self.init_content()
        
    def init_content(self):
        #Create label 
        self.gen = tk.Label(self, text="gen [kW]")
        self.dishwasher = tk.Label(self, text="Dishwasher [kW]")
        self.furnace = tk.Label(self, text="Furnace [kW]")
        self.home_office = tk.Label(self, text="Home office [kW]")
        self.fridge = tk.Label(self, text="Fridge [kW]")
        self.wine_cellar = tk.Label(self, text="Wine cellar [kW]")
        self.garage_door = tk.Label(self, text="Garage door [kW]")
        self.kitchen = tk.Label(self, text="Kitchen [kW]")
        self.barn = tk.Label(self, text="Barn [kW]")
        self.well = tk.Label(self, text="Well [kW]")
        self.microwave = tk.Label(self, text="Microwave [kW]")
        self.living_room = tk.Label(self, text="Living room [kW]")
        self.temperature = tk.Label(self, text="temperature")
        self.humidity = tk.Label(self, text="humidity")
        self.visibility = tk.Label(self, text="visibility")
        self.summary = tk.Label(self, text="summary")
        self.pressure = tk.Label(self, text="pressure")
        self.windSpeed = tk.Label(self, text="windSpeed")
        self.cloudCover = tk.Label(self, text="cloudCover")
        self.windBearing = tk.Label(self, text="windBearing")
        
        self.gen.grid(row=0, column=0, sticky=W, pady=2)
        self.dishwasher.grid(row=1, column=0, sticky=W, pady=2)
        self.furnace.grid(row=2, column=0, sticky=W, pady=2)
        self.home_office.grid(row=3, column=0, sticky=W, pady=2)
        self.fridge.grid(row=4, column=0, sticky=W, pady=2)
        self.wine_cellar.grid(row=5, column=0, sticky=W, pady=2)
        self.garage_door.grid(row=6, column=0, sticky=W, pady=2)
        self.kitchen.grid(row=7, column=0, sticky=W, pady=2)
        self.barn.grid(row=8, column=0, sticky=W, pady=2) 
        self.well.grid(row=9, column=0, sticky=W, pady=2)
        self.microwave.grid(row=10, column=0, sticky=W, pady=2) 
        self.living_room.grid(row=11, column=0, sticky=W, pady=2) 
        self.temperature.grid(row=12, column=0, sticky=W, pady=2)
        self.humidity.grid(row=13, column=0, sticky=W, pady=2)
        self.visibility.grid(row=14, column=0, sticky=W, pady=2)
        self.summary.grid(row=15, column=0, sticky=W, pady=2)
        self.pressure.grid(row=16, column=0, sticky=W, pady=2) 
        self.windSpeed.grid(row=17, column=0, sticky=W, pady=2) 
        self.cloudCover.grid(row=18, column=0, sticky=W, pady=2)
        self.windBearing.grid(row=19, column=0, sticky=W, pady=2)
        
        #create entry
        self.gen_var = tk.DoubleVar()
        self.dishwasher_var = tk.DoubleVar()
        self.furnace_var = tk.DoubleVar()
        self.home_office_var = tk.DoubleVar()
        self.fridge_var = tk.DoubleVar()
        self.wine_cellar_var = tk.DoubleVar()
        self.garage_door_var = tk.DoubleVar()
        self.kitchen_var = tk.DoubleVar()
        self.barn_var = tk.DoubleVar()
        self.well_var = tk.DoubleVar()
        self.microwave_var = tk.DoubleVar()
        self.living_room_var = tk.DoubleVar()
        self.temperature_var = tk.DoubleVar()
        self.humidity_var = tk.DoubleVar()
        self.visibility_var = tk.DoubleVar()
        self.summary_var = tk.StringVar()
        self.pressure_var = tk.DoubleVar()
        self.windSpeed_var = tk.DoubleVar()
        self.cloudCover_var = tk.DoubleVar()
        self.windBearing_var = tk.DoubleVar()
        
        self.gen_entry = tk.Entry(self, width=16, textvariable=self.gen_var)
        self.dishwasher_entry = tk.Entry(self, width=16, textvariable=self.dishwasher_var)
        self.furnace_entry = tk.Entry(self, width=16, textvariable=self.furnace_var)
        self.home_office_entry = tk.Entry(self, width=16, textvariable=self.home_office_var)
        self.fridge_entry = tk.Entry(self, width=16, textvariable=self.fridge_var)
        self.wine_cellar_entry = tk.Entry(self, width=16, textvariable=self.wine_cellar_var)
        self.garage_door_entry = tk.Entry(self, width=16, textvariable=self.garage_door_var)
        self.kitchen_entry = tk.Entry(self, width=16, textvariable=self.kitchen_var)
        self.barn_entry = tk.Entry(self, width=16, textvariable=self.barn_var)
        self.well_entry = tk.Entry(self, width=16, textvariable=self.well_var)
        self.microwave_entry = tk.Entry(self, width=16, textvariable=self.microwave_var)
        self.living_room_entry = tk.Entry(self, width=16, textvariable=self.living_room_var)
        self.temperature_entry = tk.Entry(self, width=16, textvariable=self.temperature_var)
        self.humidity_entry = tk.Entry(self, width=16, textvariable=self.humidity_var)
        self.visibility_entry = tk.Entry(self, width=16, textvariable=self.visibility_var)
        self.summary_entry = tk.Entry(self, width=16, textvariable=self.summary_var)
        self.pressure_entry = tk.Entry(self, width=16, textvariable=self.pressure_var)
        self.windSpeed_entry = tk.Entry(self, width=16, textvariable=self.windSpeed_var)
        self.cloudCover_entry = tk.Entry(self, width=16, textvariable=self.cloudCover_var)
        self.windBearing_entry = tk.Entry(self, width=16, textvariable=self.windBearing_var)
        
        self.gen_entry.grid(row=0, column=1, pady=2)
        self.dishwasher_entry.grid(row=1, column=1, pady=2)
        self.furnace_entry.grid(row=2, column=1, pady=2)
        self.home_office_entry.grid(row=3, column=1,pady=2)
        self.fridge_entry.grid(row=4, column=1, pady=2)
        self.wine_cellar_entry.grid(row=5, column=1, pady=2)
        self.garage_door_entry.grid(row=6, column=1, pady=2)
        self.kitchen_entry.grid(row=7, column=1, pady=2)
        self.barn_entry.grid(row=8, column=1, pady=2)
        self.well_entry.grid(row=9, column=1, pady=2)
        self.microwave_entry.grid(row=10, column=1, pady=2)
        self.living_room_entry.grid(row=11, column=1, pady=2)
        self.temperature_entry.grid(row=12, column=1, pady=2)
        self.humidity_entry.grid(row=13, column=1, pady=2)
        self.visibility_entry.grid(row=14, column=1, pady=2)
        self.summary_entry.grid(row=15, column=1, pady=2)
        self.pressure_entry.grid(row=16, column=1, pady=2)
        self.windSpeed_entry.grid(row=17, column=1, pady=2)
        self.cloudCover_entry.grid(row=18, column=1, pady=2)
        self.windBearing_entry.grid(row=19, column=1, pady=2)
        
        #create button
        self.button_clear = tk.Button(self, text = "Clear", command = self.clear_all)
        self.button_predict = tk.Button(self, text = "Predict", command = self.predict)
        self.button_generate = tk.Button(self, text = "Generate", command = self.generate)
        self.button_create = tk.Button(self, text= "Create", command = self.create)
        
        self.button_create.grid(row=1, column=2, pady=2)
        self.button_generate.grid(row=5, column=2, pady=2)
        self.button_predict.grid(row=9, column=2, pady=2)
        self.button_clear.grid(row=13, column=2, pady=2)
        
        #Create result entry
        self.predict = tk.Label(self, text="Predicted result", bg='#ADD8E6', font=tkFont.Font(weight='bold', size=10))
        self.predict.grid(row=20, column=0, sticky=W, pady=2)
        self.predict_value = tk.DoubleVar()
        self.predict_entrybox=tk.Entry(self, width=16, textvariable=self.predict_value)
        self.predict_entrybox.grid(row=20,column=1, pady=2)
        
    def create(self):
        #Create dataframe to store input
        global df
        df = pd.DataFrame(columns=['gen [kW]', 'Dishwasher [kW]', 'Furnace [kW]', 'Home office [kW]', 'Fridge [kW]', 
                                   'Wine cellar [kW]', 'Garage door [kW]', 'Kitchen [kW]','Barn [kW]', 'Well [kW]', 
                                   'Microwave [kW]', 'Living room [kW]', 'temperature', 'humidity', 'visibility', 
                                   'summary', 'pressure', 'windSpeed', 'cloudCover', 'windBearing'])
        
    def generate(self):
        #Generate random number
        self.gen_var.set(round(np.random.uniform(0, 1), 3))
        self.dishwasher_var.set(round(np.random.uniform(0, 1.2), 3)) 
        self.furnace_var.set(round(np.random.uniform(0, 1.4), 3))
        self.home_office_var.set(round(np.random.uniform(0, 1), 3))
        self.fridge_var.set(round(np.random.uniform(0, 1), 3))
        self.wine_cellar_var.set(round(np.random.uniform(0, 1.1), 3))
        self.garage_door_var.set(round(np.random.uniform(0, 1.1), 3))
        self.kitchen_var.set(round(np.random.uniform(0, 1.4), 3))
        self.barn_var.set(round(np.random.uniform(0, 2), 3))
        self.well_var.set(round(np.random.uniform(0, 1.2), 3))
        self.microwave_var.set(round(np.random.uniform(0, 1.3), 3)) 
        self.living_room_var.set(round(np.random.uniform(0, 0.8), 3))
        self.temperature_var.set(round(np.random.uniform(-15, 100), 3)) 
        self.humidity_var.set(round(np.random.uniform(0, 1), 3))
        self.visibility_var.set(round(np.random.uniform(0, 1001), 3))
        self.pressure_var.set(round(np.random.uniform(950, 1050), 3))
        self.windSpeed_var.set(round(np.random.uniform(0, 25), 3))
        self.cloudCover_var.set(round(np.random.uniform(0, 1.01), 3)) 
        self.windBearing_var.set(np.random.randint(0, 360))
        
        #Generate random sample for summary variable
        summary=('Clear', 'Partly Cloudy', 'Rainy', 'Snowy', 'Cloudy', 'Windy', 'Foggy')
        self.summary_var.set(np.random.choice(summary))
            
    def predict(self):
        #Get input
        gen=self.gen_var.get()
        dishwasher=self.dishwasher_var.get()
        furnace=self.furnace_var.get()
        home_office=self.home_office_var.get() 
        fridge=self.fridge_var.get() 
        wine_cellar=self.wine_cellar_var.get() 
        garage_door=self.garage_door_var.get() 
        kitchen=self.kitchen_var.get() 
        barn=self.barn_var.get() 
        well=self.well_var.get() 
        microwave=self.microwave_var.get() 
        living_room=self.living_room_var.get() 
        temperature=self.temperature_var.get() 
        humidity=self.humidity_var.get() 
        visibility=self.visibility_var.get()
        summary = self.summary_var.get()
        pressure=self.pressure_var.get() 
        windSpeed=self.windSpeed_var.get()
        cloudCover=self.cloudCover_var.get() 
        windBearing=self.windBearing_var.get()
        
        #add values to dataframe
        series_values = pd.Series({'gen [kW]': gen, 'Dishwasher [kW]': dishwasher, 'Furnace [kW]': furnace, 
                                   'Home office [kW]': home_office, 'Fridge [kW]': fridge, 'Wine cellar [kW]': wine_cellar, 
                                   'Garage door [kW]': garage_door, 'Kitchen [kW]': kitchen, 'Barn [kW]': barn, 
                                   'Well [kW]': well, 'Microwave [kW]': microwave, 'Living room [kW]': living_room, 
                                   'temperature': temperature, 'humidity': humidity, 'visibility': visibility, 
                                   'summary': summary, 'pressure': pressure, 'windSpeed': windSpeed, 'cloudCover': cloudCover, 
                                   'windBearing': windBearing})
        
        global df
        df = df.append(series_values, ignore_index=True)
            
        #Preprocess data
        X_num = df.drop('summary', axis=1)
        X_cat = df['summary']
            
        #Apply standardscaler on numerical features using training set mean and standard deviation
        X_num = sc.transform(X_num)
            
        #Define category class for summary
        value=['Clear', 'Partly Cloudy', 'Rainy', 'Snowy', 'Cloudy', 'Windy', 'Foggy']
            
        #Perform one-hot encoding
        lb_encoder = LabelBinarizer()
        lb_encoder.fit(value)
        X_cat = lb_encoder.transform(X_cat)
    
        #Combine X values
        X = np.hstack([X_num, X_cat])
            
        #Predict value
        y = mlp.predict(X)
        self.predict_value.set('{:.3f} kW'.format(float(y)))
            
    def clear_all(self):
        #Clear all input
        self.gen_var.set('')
        self.dishwasher_var.set('') 
        self.furnace_var.set('')
        self.home_office_var.set('') 
        self.fridge_var.set('') 
        self.wine_cellar_var.set('') 
        self.garage_door_var.set('') 
        self.kitchen_var.set('') 
        self.barn_var.set('') 
        self.well_var.set('') 
        self.microwave_var.set('') 
        self.living_room_var.set('') 
        self.temperature_var.set('') 
        self.humidity_var.set('') 
        self.visibility_var.set('')
        self.summary_var.set('')
        self.pressure_var.set('') 
        self.windSpeed_var.set('')
        self.cloudCover_var.set('') 
        self.windBearing_var.set('')
            
        #Clear predict result
        self.predict_value.set('')
            
        #Clear series value in dataframe
        global df
        df.drop(df.index[0], inplace=True)
```
